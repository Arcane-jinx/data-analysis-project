```python
import pandas as pd
import matplotlib.pyplot as plt

# URL del archivo CSV
url = "https://raw.githubusercontent.com/justmarkham/pandas-videos/master/data/imdb_1000.csv"

# Cargar el archivo CSV en un DataFrame
df = pd.read_csv(url)

# Mostrar las primeras filas del DataFrame
print("Primeras filas del DataFrame:")
print(df.head())

# Mostrar información general del DataFrame
print("\nInformación del DataFrame:")
print(df.info())

# Número de filas y columnas
print("\nNúmero de filas y columnas:")
print(df.shape)

# Tipos de datos de cada columna
print("\nTipos de datos de cada columna:")
print(df.dtypes)

# Presencia de valores nulos o duplicados
print("\nValores nulos en cada columna:")
print(df.isnull().sum())
print("\nNúmero de filas duplicadas:")
print(df.duplicated().sum())

# Mostrar estadísticas descriptivas
print("\nEstadísticas descriptivas:")
print(df.describe())

# Filtrar películas con una calificación (star_rating) mayor a 8.5
high_rated_movies = df[df['star_rating'] > 8.5]
print("\nPelículas con una calificación mayor a 8.5:")
print(high_rated_movies)

# Contar el número de películas por género
genre_counts = df['genre'].value_counts()
print("\nNúmero de películas por género:")
print(genre_counts)

# Visualización: Gráfico de barras del número de películas por género
plt.figure(figsize=(10, 5))
genre_counts.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Número de películas por género')
plt.xlabel('Género')
plt.ylabel('Cantidad de películas')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
